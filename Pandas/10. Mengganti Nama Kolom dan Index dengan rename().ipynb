```python
import pandas as pd
df_hasil_ujian = pd.read_csv("StudentsPerformance.csv")
df_hasil_ujian.head()
```

### Mengganti Nama Kolom

```python
# mengganti nama kolom dan menimpa nilai DataFrame
df_hasil_ujian = df_hasil_ujian.rename(columns={"gender" : "Gender"})
```

```python
# mengganti beberapa kolom sekaligus dan memperbarui DataFrame dengan parameter inplace
df_hasil_ujian.rename(columns={ "math score" : "Math",
                                "reading score": "Reading",
                                "writing score" : "Writing"
                                },inplace= True)
```

### Mengganti Nama Index

```python
# mengganti index 0, 1, dan 2 lalu memperbarui DataFrame
df_hasil_ujian.rename(index={ 0:"A", 1:"B", 2:"C"}, inplace=True)
```
