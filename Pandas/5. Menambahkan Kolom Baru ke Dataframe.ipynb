## 2. Menambahkan Kolom Baru dengan menggunakan assign() dan insert()

### 2.1 assign()

Kapan sebaiknya menggunakan assign()?

Untuk menambahkan beberapa kolom sekaligus hanya dalam satu baris kode.

Ketika kamu perlu menimpa (overwrite) nilai dari kolom yang sudah ada (best practice).

assign() akan mengembalikan objek baru (salinan) yang berisi semua kolom asli ditambah dengan kolom baru.

```python
# membuat angka acak untuk kolom skor baru
score1 = np.random.randint(1,100,size=1000)
score2 = np.random.randint(1,100,size=1000)
```

```python
# membuat sebuah Series menggunakan angka acak
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
# menggunakan assign() untuk menambahkan beberapa kolom sekaligus
df_hasil_ujian = df_hasil_ujian.assign(score1=series1, score2=series2)
```

```python
df_hasil_ujian
```

### 2.2 insert()
Menyisipkan kolom baru pada posisi atau indeks tertentu.

```python
# menggunakan insert() untuk menambahkan kolom pada posisi tertentu
df_hasil_ujian.insert(1,"test",series1)
```
