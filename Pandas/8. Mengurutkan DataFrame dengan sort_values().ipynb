```python
import pandas as pd

# membaca dan menampilkan DataFrame
df_nilai_siswa = pd.read_csv("StudentsPerformance.csv")
df_nilai_siswa.head()
```

### Mengurutkan sebuah DataFrame

```python
# mengurutkan berdasarkan satu kolom
df_nilai_siswa.sort_values("math score")
```

```python
# mengurutkan descending berdasarkan satu kolom
df_nilai_siswa.sort_values("math score", ascending=False)
```

```python
# mengurutkan menurun berdasarkan beberapa kolom
df_nilai_siswa.sort_values(["math score", "reading score"], ascending=False)
```

```python
# mengurutkan menurun berdasarkan beberapa kolom dan memperbarui DataFrame
df_nilai_siswa.sort_values(["math score","reading score"], ascending=True, inplace=True)
df_nilai_siswa
```

```python
# mengurutkan menurun dengan key function (biasanya dipake saat mengurutkan string)
df_nilai_siswa.sort_values("race/ethnicity",ascending=True, key=lambda a:a.str.lower())
```
