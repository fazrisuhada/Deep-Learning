```python
import pandas as pd
df_nilai_siswa = pd.read_csv("StudentsPerformance.csv")
df_nilai_siswa.head()
```

### Membuat Index

```python
import numpy as np
import random
```

```python
# membuat nilai-nilai unik untuk index
new_index = np.arange(0,1000)
new_index
```

```python
# mengacak urutan index
random.shuffle(new_index)
new_index
```

```python
# membuat kolom baru dengan index baru
df_nilai_siswa["new_index"] = new_index
df_nilai_siswa
```

### Menetapkan Index

```python
# menetapkan kolom new_index sebagai index
df_nilai_siswa.set_index('new_index', inplace=True)
df_nilai_siswa
```

### Mengurutkan Index

```python
# mengurutkan DataFrame berdasarkan index baru
df_nilai_siswa.sort_index()
```

```python
# mengurutkan index secara menurun
df_nilai_siswa.sort_index(ascending=False)
```
